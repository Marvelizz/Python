```python
##### Q.program to find occurrence of letters in a string
```

```python
inp_str=input("Enter the string :")

count = {}

for a in inp_str:
    if a in count:
        count[a] += 1
    else:
        count[a] = 1
print ("Occurrence of each character :")
for a,count in count.items():
    print (f"{a}: {count}")
```

```python
#### Q.Create a calculator using functions??
```

```python
def add(x,y):
  return x+y
def mult(x,y):
  return x*y
def div(x,y):
  if y!=0:
   return x/y
  else:
    "cannot divided by zero"
def sub(x,y):
  return x-y
while True:
  print("Select option.")
  print("1.Add")
  print("2.Multiply")
  print("3.Substract")
  print("4.Divide")
  print("5.Exit")

  choice=input("Enter a choice (1/2/3/4/5) :")

  if choice=='5':
    print("Exiting the calculator.")
    break
  elif choice>'5':
    print("Invalid choice")
    break
  num1=float(input("Enter 1st number :"))
  num2=float(input("Enter 2nd number :"))
  if choice=='1':
   print("Result :",add(num1,num2))
  elif choice=='2':
   print("Result :",mult(num1,num2))
  elif choice=='3':
   print("Result :",sub(num1,num2))
  elif choice=='4':
   print("Result :",div(num1,num2))
```
